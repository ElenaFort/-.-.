**Урок 10. Дисперсионный анализ**

Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.
Даны значения роста в трех группах случайно выбранных спортсменов:

Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.

Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.

Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

x1 = np.array([173, 175, 180, 178, 177, 185, 183, 182])
x2 = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
x3 = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])

n1 = x1.shape[0]
n2 = x2.shape[0]
n3 = x3.shape[0]

k = 3
n = n1 + n2 + n3

x_mean_1 = np.mean(x1)
x_mean_2 = np.mean(x2)
x_mean_3 = np.mean(x3)

x_mean_1, x_mean_2, x_mean_3
```

```python
total = np.concatenate([x1, x2, x3])
total
```

```python
x_mean_total = np.mean(total)
x_mean_total
```

```python
#Сумма квадратов отклонений наблюдений от общего среднего

np.sum((total - x_mean_total)**2)
```

```python
#Сумма квадратов отклонений средних групповых значений от общего среднего

S_f = ((x_mean_1 - x_mean_total)**2)*n1 + ((x_mean_2 - x_mean_total)**2)*n2 + ((x_mean_3 - x_mean_total)**2)*n3
S_f
```

```python
#Остаточная сумма квадратов отклонений

S_ost = np.sum((x1 - x_mean_1)**2) + np.sum((x2 - x_mean_2)**2) + np.sum((x3 - x_mean_3)**2)
S_ost
```

```python
S_f + S_ost
```

```python
D_f = S_f / (k - 1)
D_f
```

```python
D_ost = S_ost / (n - k)
D_ost
```

```python
F_n = D_f / D_ost
F_n
```

```python
f = stats.f_oneway(x1, x2, x3)
f
```

```python
#Степени свободы числителя
k - 1
```

```python
#Степени свободы знаменателя
n - k
```

Фишер.png

Расчетное значение 5.5 больше табличного значения, а значит попадает в область принятия альтернативной гипотезы. Т.е. мы делаем вывод, что средний рост футболистов, хоккеистов и штангистов имеет статистически значимые различия.

Так же мы можем сделать выводы по значению pvalue = 0.01 < 0.05, значит принимаем альтернативную гипотезу, средний рост имеет статистически значимые различия.
